Q] What do you mean by gray level

A. The grey level or grey value indicates the brightness of a pixel. The minimum grey level is 0. The maximum grey level depends on the digitisation depth of the image. For an 8-bit deep image it is 255. In a binary image a pixel can only take on either the value 0 or the value 255.
