```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
cat = cv2.imread('/content/cat1.jpg')
```

```python
plt.imshow(cat)
```

```python
show_cat = cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_cat)
```

```python
img_gray = cv2.imread('cat1.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
hist_values = cv2.calcHist([img_gray],channels = [0], mask=None, histSize = [256], ranges = [0, 256])
plt.plot(hist_values)
```

**Histogram Equalization**

```python
eq_cat = cv2.equalizeHist(img_gray)
plt.imshow(eq_cat, cmap = 'gray')
```

```python
hist_values_eq = cv2.calcHist([eq_cat],channels = [0], mask=None, histSize = [256], ranges = [0, 256])
plt.plot(hist_values_eq)
```

```python
fig = plt.figure(dpi=300)
plt.subplot(221), plt.imshow(img_gray, cmap='gray'), plt.title('Original Image')
plt.subplot(222), plt.imshow(eq_cat, cmap='gray'), plt.title('Equalized Image')
plt.subplot(223), plt.plot(hist_values), plt.title('Plot of Original Image')
plt.subplot(224), plt.plot(hist_values_eq), plt.title('Plot of Equalized Image')
```

```python
img = cat
color = ('b', 'g', 'r')
for i, col in enumerate(color):
  histr = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(histr,color = col)
  plt.xlim([0, 256])
plt.title('Cat Image')
plt.show()
```

**Masking**

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
mask[40:100, 120:200] = 255
```

```python
plt.imshow(mask, cmap='gray')
```

```python
masked_img = cv2.bitwise_and(img, img, mask = mask)
show_masked_img = cv2.bitwise_and(cat, cat, mask=mask)
```

```python
plt.imshow(show_masked_img)
```

```python
img2 = show_masked_img
color = ('b', 'g', 'r')
for i, col in enumerate(color):
  histr = cv2.calcHist([img2],[i],None,[256],[0,256])
  plt.plot(histr,color = col)
  plt.xlim([0, 256])
plt.title('Masked Image')
plt.show()
```

```python
fig = plt.figure(dpi=300)
plt.subplot(221), plt.imshow(img_gray, cmap='gray'), plt.title('Original Image')
plt.subplot(222), plt.imshow(show_masked_img, cmap='gray'), plt.title('Masked Image')
plt.subplot(223), plt.plot(hist_values), plt.title('Plot of Original Image')
plt.subplot(224), plt.plot(histr), plt.title('Plot of Masked Image')
```

### PostLab Questions 

1. A bright image will have what kind of histogram?
- A bright image will have the histogram whose component are biased toward high side of gray scale. A high contrast image will have the histogram that covers wide range of gray scale and the distribution of pixel is approximately uniform.

2. What is ‘Dynamic range’ and ‘Contrast’ of digital images?
- Dynamic range describes the ratio between the brightest and darkest parts of an image, from pure black to brightest white. The term contrast refers to the amount of color or grayscale differentiation that exists between various image features in both analog and digital images. Images having a higher contrast level generally display a greater degree of color or grayscale variation than those of lower contrast.
